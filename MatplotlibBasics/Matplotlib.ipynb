### Plots Using Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,7,9,10,16])
plt.plot(x)
plt.show()
```

```python
x = np.array([1,2,3,4,7,9,10,16])
plt.plot(x, x**2)
plt.plot(x, x**3)
plt.show()
plt.plot(x, np.sin(x))
plt.show()
```

```python
y = np.array([1,2,3,4,5,6,7,8])
plt.scatter(x, y, color="Red", marker = "^")
plt.title("Stock Data", color="red")
plt.xlabel("Days")
plt.ylabel("Price")
plt.show()
```

```python
x = np.random.random((1000, 2))
for ix in range(x.shape[0]):
    plt.scatter(x[ix, 0], x[ix, 1], marker = '+', color = "Blue")
plt.show()
```
